```python
from trading import *
from pytrends.request import TrendReq

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pytrends = TrendReq(hl="en-IN", tz=360)
```

```python
query = "Reliance"
kw_list = [query]
pytrends.build_payload(kw_list, cat=0, timeframe="all", geo="", gprop="")
```

```python
ticker = YTicker("RELIANCE.NS")
plot = plotting.LinePlot(ticker)
plot.plot(period=periods.max, interval=intervals.day)

ax = plt.gca()
ax = ax.twinx()
ax.plot(pytrends.interest_over_time()[query], color="red")
```
