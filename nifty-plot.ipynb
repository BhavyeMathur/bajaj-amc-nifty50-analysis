```python
NIFTY = YTicker("^NSEI")
NIFTY.print_info()
```

```python
plot = plotting.CandlestickPlot(NIFTY)

plot.indicator(BollingerBands())
plot.indicator(SMAIndicator(window=20))

plot.show(period=periods.ytd, interval=intervals.day)
```

```python
plot = plotting.LinePlot(NIFTY)
plot.show(period=periods.max, interval=intervals.day)
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

data = NIFTY.history(period=periods.max, interval=intervals.day)["% Change"]
data.dropna(inplace=True)
data.hist(bins=250, density=True, grid=False)

x = np.linspace(*plt.xlim(), 100)
p = norm.pdf(x, *norm.fit(data))

plt.plot(x, p, color="#000", linewidth=1)
```
